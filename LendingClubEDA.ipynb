```
# Cleaning data using Spark UDF.
from pyspark.sql.types import FloatType

def trim(string):
  return string.strip('%')

# we register the above python function as a UDF
spark.udf.register('trimpct',trim)
```

```
%sql
select int_rate, cast(trimpct(int_rate) as float) as int_rate_float from loan_stats_final 
```

```
%sql

select bad_loan, cast(trimpct(int_rate) as float) as int_rate_float from loan_stats_final 
```
